# Energy momentum conservation for phonon scattering

General energy constraint for phonon matching to a spherical parabolic band, once momentum is conserved [Eq 2.52 in Lundstrom]:

$\hbar\beta=2p \left[ \mp\cos\theta \pm \frac{m^*\omega}{p\beta} \right]$

where $\beta$ is the phonon wavevector magnitude, and $\omega$ is the phonon energy, and the top-sign is for absorption, bottom for emission.
The maximum wavevector for absorption occurs when $\theta=\pi$, and for emission when $\theta=0$

$\hbar\beta_\mathrm{max}=2p \left[ 1 \pm \frac{m^*\omega_\mathrm{max}}{p\beta} \right]$

## Acoustic Phonons
Plugging in an zone-center acoustic dispersion:

$\hbar\beta_\mathrm{max}=2p \left[ 1 \pm \frac{m^*v_s}{p} \right]$

For 2DEG mobility, we'll be most interested in the wavevector-in-plane LA phonon, which, from the [Bulk Acoustic Phonons notebook](https://github.com/samueljamesbader/PyNitride/
tree/examples/Bulk_Acoustic_Phonons.ipynb), has a sound velocity of $v_s=8\times 10^5$cm/s.  Let's examine the scale of the last term, assuming the GaN 2DEG is degenerate by about 100meV.

```python
import pynitride as pn
pmdb=pn.ParamDB(units='Pint')
parse=pmdb.parse
to_units=pmdb.to_units
hbar=pmdb.get_constants("hbar")
import numpy as np
```

```python
hbar
```

```python
GaN=pn.Material("GaN",pmdb=pmdb)
m=GaN['electron.mxys']
vs=parse("8e5 cm/s")
```

```python
p=np.sqrt(2*m*parse("100meV"))
vt=p/m
print("Fermi velocity in GaN is {:.2g} cm/s".format(to_units(vt,"cm/s")))
```

So the carrier velocity is way larger than the sound velocity, and the last term will be small.  So, $\hbar \beta_\mathrm{max}\approx 2p$

```python
beta_max=2*p/hbar

print("Max wavevector is 2pi / {:.2g} nm".format(to_units(2*np.pi/beta_max,"nm")))
```

For comparison, the K-point of the Brillouin zone in GaN is at 

```python
print("K point wavevector magnitude 2pi / {:.2g} nm".format(to_units(2*np.pi/(4*np.pi/(3*GaN['lattice.a'])),"nm")))
```

So the allowable acoustic phonons for scattering are close to zone-center

```python
print("The maximum scattering acoustic phonon energy is {:.2g} meV".format(to_units(hbar*vs*beta_max,"meV")))
```

## Optical Phonons

Plugging in the optical (constant) dispersion

$\hbar\beta_\mathrm{max}=2p \left[ 1 \pm \frac{m^*\omega_\mathrm{OP}}{p\beta_\mathrm{max}} \right]$

which leads to 

$\hbar \beta_\mathrm{max}=p\left[1+\sqrt{1\pm \frac{\hbar\omega_\mathrm{OP}}{E(p)}}\right]$

where $E(p)$ is the initial electron energy

Optical phonons in GaN are about 91meV.  So for a Fermi level of about 100meV then, the max phonon momentum is about 2.5-3 times $p$ for absorption, and about $p$ for emission.  So again, we're talking about near-zone-center phonons
