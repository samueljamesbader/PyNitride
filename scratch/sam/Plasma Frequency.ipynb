```python
import pynitride as pn
pmdb=pn.ParamDB(units='Pint')
parse=pmdb.parse
to_units=pmdb.to_units
h,hbar=pmdb.get_constants("h,hbar")
import numpy as np
```

```python
q=parse("2*pi/4nm")
n02d=parse(".5e13/cm^2")
n03d=parse("1e18/cm^3")
e=parse("e")
m=.2*parse("m_e")
eps_0=parse("epsilon_0")
eps=pn.Material("GaN",pmdb=pmdb)['dielectric.eps']
eps_inf=5*eps_0
```

```python
(hbar*np.sqrt(n02d*e**2/(2*eps_inf*m*q))).to("meV")
```

```python
np.sqrt(n02d*e**2/(2*eps_inf*m*q)).to("THz")
```
