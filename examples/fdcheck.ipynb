```python
%matplotlib notebook
from matplotlib.pyplot import *
import scipy
pi=scipy.pi
from pynitride.poissolve.maths import fd12
```

```python
def H_fd12(etac):
 #pierre
 return 1/(scipy.exp(-etac)+3*scipy.sqrt(pi/2)*
                     ((etac+2.13)+(abs(etac-2.13)**2.4+9.6)**(5./12))**(-1.5))
```

```python
H_fd12(0)
```

```python
fd12(0)
```

```python
x=scipy.linspace(-10,1000)
```

```python
figure()
plot(x,fd12(x),'.')
plot(x,H_fd12(x),)
#yscale('log')
```
