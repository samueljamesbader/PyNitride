GaN HEMT
===

```python
mpl.figure()
hemt=gan_hemt(.001*nm,20*nm,500*nm,1e16*cm**-3,surface=1*eV)
Coupled_Schrodinger_Poisson(hemt,carriers=['electron','hole']).solve(rise=100)#low_act=5,rise=200,callback=stoppah)
plot_wavefunctions(hemt)

mpl.xlim(0,40)
mpl.sca(mpl.gcf().get_axes()[0])
mpl.ylim(-.5,.5)
mpl.hlines(y=0,xmin=0,xmax=100,color='r',linewidth=2)
```

```python
hemt['n'].integrate()[-1]/(1e13/cm**2)
```
